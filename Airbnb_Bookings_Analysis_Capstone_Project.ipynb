# Data Cleaning & Finding Missing Data

```python
# finding the total number of missing data in each columns
df.isnull().sum()
```


> Found the 'last_review' and 'reviews_per_month' columns have 10052 missing values each.



> Missing data that has been observed does not require any special treatment. Further, given the nature of our dataset, we can state that columns "name", "host_name" and "laast_review" are irrelevant and insignificant to our data analysis and considering the privacy and ethical issues we drop these columns, while columns "review per month" require very simple handling. For "review per month" we can add 0.0 for missing values.



```python
# Dropping the non relevant columns
df.drop(['host_name','last_review','name'],axis=1,inplace=True)
```

```python
# Replacing missinng values in reviews_per_month column with zero
df["reviews_per_month"].fillna(0, inplace = True)
```
